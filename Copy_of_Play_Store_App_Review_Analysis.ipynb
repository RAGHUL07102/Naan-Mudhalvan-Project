```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Training_Data_Google_Play_reviews_6000.csv')
```

```python
# Display basic information about the dataset
print("Number of rows and columns:", data.shape)
print("\nColumns present in the dataset:")
print(data.columns)
```

```python
# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(data.head())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(data.describe())
```

```python
# Check for missing values
print("\nMissing values:")
print(data.isnull().sum())
```

```python
# Analyze ratings
print("\nAverage rating:", data['score'].mean())
print("Maximum rating:", data['score'].max())
print("Minimum rating:", data['score'].min())
```

```python
# Analyze thumbs up count
print("\nAverage thumbs up count:", data['thumbsUpCount'].mean())
print("Maximum thumbs up count:", data['thumbsUpCount'].max())
print("Minimum thumbs up count:", data['thumbsUpCount'].min())
```

```python
# Analyze review creation version
print("\nNumber of unique review creation versions:", data['reviewCreatedVersion'].nunique())
```

```python
# Analyze app versions
print("\nNumber of unique app versions:", data['appVersion'].nunique())
```

```python
# Analyze user languages
print("\nNumber of unique user languages:", data['userLang'].nunique())
```

```python
# Visualization
# Distribution of ratings
plt.figure(figsize=(8, 6))
sns.histplot(data['score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of thumbs up count
plt.figure(figsize=(8, 6))
sns.histplot(data['thumbsUpCount'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Thumbs Up Count')
plt.xlabel('Thumbs Up Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Barplot of top 10 categories by review count
top_categories = data['content'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top 10 Categories by Review Count')
plt.xlabel('Category')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.show()
```
